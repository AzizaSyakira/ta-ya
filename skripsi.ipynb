```python
import pandas as pd

df = pd.read_excel('data bps.xlsx', header=None)
years = df.iloc[0]
months = df.iloc[1]
values = df.iloc[2]
bulan_map = {
    'Januari': '01', 'Februari': '02', 'Maret': '03', 'April': '04',
    'Mei': '05', 'Juni': '06', 'Juli': '07', 'Agustus': '08',
    'September': '09', 'Oktober': '10', 'November': '11', 'Desember': '12'
}
data = pd.DataFrame({
    'Year': years,
    'Month': months,
    'Visitors': values
})
data = data.dropna()
data['Month'] = data['Month'].map(bulan_map)
data['Date'] = pd.to_datetime(
    data['Year'].astype(int).astype(str) + '-' + data['Month'] + '-01'
)
df_final = data[['Date', 'Visitors']].sort_values('Date')
df_final.set_index('Date', inplace=True)
df_final.head()
```

```python
from pytrends.request import TrendReq
```

```python
kw_pariwisata = [
    "wisata sumatera barat",
    "liburan ke padang",
    "wisata bukittinggi",
    "tiket pesawat padang",
    "hotel di padang"
]

kw_pendidikan = [
    "kuliah di padang",
    "universitas di sumatera barat",
    "wisuda unand",
    "kost di padang"
]

kw_adat = [
    "pulang basamo",
    "mudik lebaran",
    "tradisi minangkabau",
    "lebaran di padang"
]
```

```python
pytrends = TrendReq(hl='id-ID', tz=360)

def get_related_queries(keyword):
    pytrends.build_payload([keyword], geo='ID')
    related = pytrends.related_queries()

    top = related[keyword]['top']
    rising = related[keyword]['rising']

    return top, rising
```

```python
top, rising = get_related_queries("wisata sumatera barat")

print(top)
print(rising)
```

```python
def collect_related(keywords):
    all_kw = []

    for kw in keywords:
        top, rising = get_related_queries(kw)

        if top is not None:
            all_kw.extend(top['query'].tolist())

        if rising is not None:
            all_kw.extend(rising['query'].tolist())

    return list(set(all_kw))
```

```python
kw_pariwisata_all = collect_related(kw_pariwisata)
kw_pendidikan_all = collect_related(kw_pendidikan)
kw_adat_all = collect_related(kw_adat)

print(len(kw_pariwisata_all))
print(len(kw_pendidikan_all))
print(len(kw_adat_all))
```
